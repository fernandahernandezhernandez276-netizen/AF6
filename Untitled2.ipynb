## Construcción del Modelo CNN

```python
model = models.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')
])

model.summary()
```

```python
# Celda 3: Normalización de imágenes
train_data = train_data / 255.0
test_data = test_data / 255.0

# Opcional: Ver la forma de los datos para tu reporte
print(f"Datos de entrenamiento: {train_data.shape}")
print(f"Datos de prueba: {test_data.shape}")
```

```python
# Celda 4: Definición de la estructura de la red
model = models.Sequential([
    layers.Flatten(input_shape=(28, 28)),  # Aplana la imagen de 28x28
    layers.Dense(128, activation='relu'),   # Capa oculta
    layers.Dense(10, activation='softmax')  # Capa de salida requerida
])

# Compilación del modelo (Definir optimizador y función de pérdida)
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Entrenamiento del modelo usando el método .fit()
history = model.fit(train_data, train_labels, epochs=5, validation_split=0.1)
```

```python
# Celda 5: Evaluar la exactitud con el conjunto de prueba
test_loss, test_acc = model.evaluate(test_data, test_label, verbose=2)
print(f'\nExactitud en el conjunto de prueba (Accuracy): {test_acc:.4f}')
```

```python
# Celda 6: Realizar 5 predicciones y mostrar el valor real vs predicción
predictions = model.predict(test_data)

# Tomamos las primeras 5 imágenes del conjunto de prueba
for i in range(5):
    prediccion_final = np.argmax(predictions[i])
    valor_real = test_label[i]

    print(f"Muestra {i+1}: Predicción del modelo = {prediccion_final} | Valor real = {valor_real}")
```
